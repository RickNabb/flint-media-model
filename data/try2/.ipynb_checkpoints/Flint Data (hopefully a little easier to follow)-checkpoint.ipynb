```python
import csv
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
'''TASK 1: PLOTTING THE BEST CASE RUNS'''
```

```python
#the csv below provides the top 50 runs per the entire influence model sweep

best_150_infl_sweep=pd.read_csv('best_50_cases_per_metric_inf_sweep.csv')
best_50_runs_rmse=best_150_infl_sweep.loc[best_150_infl_sweep["metric"] =='rmse']
#print(best_50_runs_rmse

#load google data
g_data=pd.read_csv("google-trends.csv")
g_data.columns = ['week', 'data']
goog_data = g_data['data'].to_list()
#print(goog_data)

def lighten_color(color, amount):
    """
    Lightens the given color by multiplying (1-luminosity) by the given amount.
    Input can be matplotlib color string, hex string, or RGB tuple.

    Examples:
    >> lighten_color('g', 0.3)
    >> lighten_color('#F034A3', 0.6)
    >> lighten_color((.3,.55,.1), 0.5)
    """
    import matplotlib.colors as mc
    import colorsys
    try:
        c = mc.cnames[color]
    except:
        c = color
    c = colorsys.rgb_to_hls(*mc.to_rgb(c))
    return colorsys.hls_to_rgb(c[0], 1 - (amount) * (1 - c[1]), c[2])

    
time_list=[]
for i in range(0, 114):
    time_list.append(i)
    

for i in range(0, 25):
    data=best_50_runs_rmse['short list'].iloc[i]
    data_list = data.strip('][').split(', ')
    new_list=[]
    #print(data)
    #print(data_list)
    #print(max(data_list))
    for i in range(0, len(data_list)):
        item=float(data_list[i])
        new_list.append(item)
    #print(max(new_list))
    plt.tick_params(left=True, right = False , labelleft = True ,
                labelbottom = True, bottom = False)
    val=i
    plt.plot(time_list, new_list, color='red')
    #plt.show()
    
    
    #print(data)

plt.plot(time_list, goog_data, color='blue', lw=1, label='google')
plt.legend()
plt.title("25 Best Fits According to Pearson's RMSE")
#cmap = plt.colormaps['blues']
plt.show()

```
